```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
ms = MinMaxScaler()
```

```python
  x = data.drop(columns = ["Zone 3  Power Consumption"])
  y = data[["Zone 3  Power Consumption"]]
```

```python
x = np.array(x)
y = np.array(y)
x = ms.fit_transform(x)
y = ms.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
model = Sequential()
model.add(Dense(50,input_dim = x_train.shape[1],activation = "elu"))
model.add(Dense(30,activation='elu'))
model.add(Dropout(0.2))
model.add(Dense(30,activation='elu'))
model.add(Dropout(0.2))
model.add(Dense(20,activation='elu'))
model.add(Dense(1,activation='elu'))
```

```python
model.summary()
```

```python
model.compile(loss = "mse", optimizer = "SGD",metrics = ["mse"])
```

```python
model.fit(x_train, y_train,epochs=50, batch_size = 20, verbose = 1)
```

```python
  pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
pred = np.squeeze(pred)
y_test = np.squeeze(np.array(y_test))
r2_score(y_test,pred)
```
